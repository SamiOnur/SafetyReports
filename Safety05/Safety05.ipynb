# 1. Introduction

Numerous studies indicate that alloy characteristics like as composition, strain rate, microstructure, and service temperature may affect mechanical qualities and failure mechanisms. We can infer that the rate of deformation (strain rate) for those components of the vehicles is high in light of the dynamic energy absorption during the collision. This research examines the stress-strain behavior of aluminum 7075 T6 temper, which has the following composition: 90.0% Al, 5.6% Zn, 2.5% Mg, 0.23 % Cr, and 1.6% Cu. One of the strongest aluminum alloys on the market, this one is useful in high-stress circumstances. Gears and shafts, missile components, regulating valve components, worm gears, aerospace, and automotive applications are among the major uses. The RSO rotary hammer dynamic testing stand was used for the tests. We will be able to determine parameters like true stress, true strain, and by adding strain rate obtain relationship between these values (D, p parameters) depicted by surface graph by evaluating data on the stress-strain behavior of this aluminum alloy during the plastic flow region of the graph. employed the Cowper-Symonds strengthening model.

# 2. Calculations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data.xlsx')
```

```python
df.head(3)
```

```python
df["True_Strain"] = np.log(1+df["strain eng"])
```

```python
df["True_Stress"] = df["stress eng"]*(1+df["strain eng"])
```

```python
df.head(3)
```

```python
df["Dynamic_Stress"] = df["True_Stress"]*(1+((df["strain rate"]/0.0214)**(1/6.3096)))
```

```python
df.head(3)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(df['strain rate'], df['True_Stress'], df['Dynamic_Stress'], cmap='viridis')
ax.set_xlabel('Strain Rate')
ax.set_ylabel('Static Stress')
ax.set_zlabel('Dynamic Stress')
plt.show()
```
